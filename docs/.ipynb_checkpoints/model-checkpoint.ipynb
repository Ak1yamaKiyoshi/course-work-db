# Бізнес-план піцерії

## 1. Огляд бізнес-моделі
Піцерія спеціалізується на приготуванні та доставці піци.

Ключові елементи бізнес-моделі:
- Меню з класичних та фірмових піц
- Доставка замовлень та самовивіз
- Система регулярних акцій
- Можливість онлайн-замовлення через сайт

## 2. Бізнес-процеси

### Каталог продукції
Меню включає наступні категорії:
- Класичні піци - традиційні рецепти
- Фірмові піци - авторські варіації
- Конструктор піци - створення власної піци
- Напої - безалкогольні та алкогольні

Кожна позиція в меню представлена:
- Назвою
- Складом
- Зображенням
- Ціною
- Розміром

### Система топінгів
Топінги - це додаткові інгредієнти, які клієнт може додати до своєї піци за додаткову плату:
- Кожен топінг має власну ціну та категорію (м'ясні, сирні, овочеві, інші)
- Клієнт може додавати необмежену кількість різних топінгів до піци
- Деякі топінги можуть бути включені як стандартні для певних видів піц
- Система дозволяє клієнтам персоналізувати замовлення відповідно до їх смаку
- Топінги створюють додаткове джерело доходу та підвищують середній чек

### Процес оформлення замовлення
Для всіх клієнтів доступно:
- Обрання піци з меню
- Додавання топінгів за бажанням
- Додавання піци в кошик
- Замовлення піци
- Відстеження статусу замовлення

Для зареєстрованих клієнтів:
- Автоматичне заповнення контактних даних

Для нових клієнтів:
- Необхідність введення імені та номера телефону

### Система управління замовленнями
Після оформлення замовлення воно проходить через такі етапи:
- "Замовлення прийнято" — замовлення зареєстроване в системі
- "Замовлення готується" — піца в процесі приготування
- "Замовлення в дорозі" — кур'єр доставляє замовлення

Відстеження статусу замовлення доступне всім клієнтам через cookies браузера.

### Система лояльності
Програма лояльності включає:
- Регулярні акції — щотижневі знижки на окремі позиції в меню
- Комбо-набори — спеціальні пропозиції "піца + напій" за зниженою ціною
- Сезонні пропозиції — тематичні піци відповідно до сезону
- Спеціальні пропозиції топінгів - періодичні знижки на певні категорії топінгів

## 3. Взаємодія з клієнтами

### Обслуговування у фізичній точці
Фізична точка доступна лише для самовивозу замовлень.

### Онлайн-сервіс
Веб-сайт забезпечує:
- Перегляд меню піцерії
- Вибір та персоналізацію піци з топінгами
- Оформлення замовлення
- Відстеження статусу замовлення
- Можливість реєстрації

## 4. Технічна реалізація

### Веб-сайт
Функціональність веб-сайту:
- Відображення каталогу піц
- Інтерактивна система вибору топінгів
- Система онлайн-замовлення
- Система відстеження статусу замовлення через cookies
- Форма реєстрації для збереження контактних даних

### Обробка замовлень
- Передача замовлень на кухню
- Формування черги замовлень
- Контроль статусу виконання
- Організація доставки

## 5. Висновки
Бізнес-модель піцерії включає онлайн-замовлення, приготування піци з можливістю персоналізації через систему топінгів, та її доставку або самовивіз. Зареєстровані користувачі отримують перевагу автоматичного заповнення контактних даних.

Система відстеження замовлень доступна всім клієнтам незалежно від реєстрації. Регулярні акції сприяють підвищенню продажів та залученню нових клієнтів. 

Система топінгів дозволяє збільшити прибутковість бізнесу, даючи клієнтам можливість персоналізувати свої замовлення та одночасно підвищуючи середній чек замовлення. Ця функція також збільшує задоволеність клієнтів, оскільки вони отримують піцу, що повністю відповідає їхнім смаковим уподобанням.
